```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos los parámetros
alpha = 10.0    # Número de Prandtl
beta = 28.0     # Número de Rayleigh
sigma = 8/3     # Relación longitud-altura

# Definimos el sistema de ecuaciones diferenciales de Lorenz
def lorenz_system(t, variables):
    x, y, z = variables
    dx_dt = alpha * (y - x)
    dy_dt = x * (beta - z) - y
    dz_dt = x * y - sigma * z
    return [dx_dt, dy_dt, dz_dt]

# Condiciones iniciales
initial_conditions = [1.0, 1.0, 1.0]

# Intervalo de tiempo
t_span = (0, 40)  # de 0 a 40 segundos
t_eval = np.linspace(0, 40, 10000)  # Tiempo para evaluar la solución

# Resolver el sistema usando el integrador de Runge-Kutta de 4to orden
solution = solve_ivp(lorenz_system, t_span, initial_conditions, t_eval=t_eval)

# Graficar la solución
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution.y[0], solution.y[1], solution.y[2])
ax.set_title("Sistema de Lorenz - Atractor Caótico")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
